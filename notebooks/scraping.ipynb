```python
from bs4 import BeautifulSoup
import re as re
import time
import pandas as pd
import os
import numpy as np
```

```python
# 1. Initialize the driver with options
driver = webdriver.Chrome()
url = "https://www.nj.com/entertainment/2018/04/ranking_the_50_best_characters_in_the_marvel_cinem.html"
driver.get(url)


options = webdriver.ChromeOptions()
options.add_argument("--incognito")


# 2. Scrolling down
start = time.time()
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    end = time.time()
    if end - start > 60:
        break

# 3. Getting elements
html = driver.execute_script("return document.body.outerHTML;")
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
list_characters = soup.find_all("b")
```

```python
list_characters[1].getText().strip()
```

```python
top_characters = [i.getText().strip().upper() for i in list_characters]
top_characters.remove('BY JEREMY SCHNEIDER | NJ ADVANCE MEDIA FOR NJ.COM')
```

```python
top_characters = pd.DataFrame(top_characters)
```

```python
top_characters['numberCharacter'] = top_characters
```

```python
top_characters['numberCharacter'].replace({"29. J.A.R.V.I.S.":"29. JARVIS",
                                           "15. DR. STRANGE":"15. DR STRANGE", 
                                           "34. M'BAKU":"34. M BAKU",
                                           "19. ANT-MAN":"19. ANT MAN",
                                           "5. SPIDER-MAN":"5. SPIDER MAN"
                                           
                                          }, inplace=True)
```

```python
top_characters
```

```python
top_characters["character"] = top_characters.numberCharacter.str.extract('([a-zA-Z\s]+)')
```

```python
top_characters.drop(["numberCharacter"], axis=1, inplace=True)
```
